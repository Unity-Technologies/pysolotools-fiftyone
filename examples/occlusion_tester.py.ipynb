```python
import numpy as np
```

```python
def _instances_to_palette(color_map):

    print(f'instances to palette')

    palette_data = np.full(256 * 3, 0, dtype=np.uint8)
    id_to_index_map = {}
    curr = 0

    if len(color_map) > 256:
        print(f"Occlusion heatmap only supports up to 256 unique instances, received: {len(color_map)}")
        return palette_data

    print(f'instances length: {len(color_map)}')

    for i in color_map:
        print(f'i: {i}')
        id_to_index_map[i] = curr
        idx = curr * 3
        print(f'idx: {idx}')
        print(f'color: {color_map[i][0:3]}')

        palette_data[idx: idx + 3] = color_map[i][0:3]
        curr += 1

    return palette_data, id_to_index_map
```

```python
def _create_palettized_image(palette):
    p_img = PIL.Image.new('P', (16, 16))
    p_img.putpalette(palette)
    return p_img
```
